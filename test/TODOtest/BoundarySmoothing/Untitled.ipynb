```python
import trimesh as tri
import numpy as np
from analysis_code.lib.utils import *
import matplotlib.pyplot as plt
from analysis_code.lib.Mesh import *
from matplotlib import cm
```

```python
d = read_dat_gen("projected.out")
m = np.zeros((1000,1000))
for data in d:
    index1,index2 = int(data[0]), int(data[1])
    m[index1,index2] = data[3]
m = m.T
```

```python
fig = plt.figure(dpi=300)
ax  = fig.add_subplot(111)
p = cm.ScalarMappable(cmap=cm.jet)
p.set_array([0,0.6])
p.set_clim(0,0.6)

colors = p.to_rgba(m)
for i in range(1000):
    for j in range(1000):
        if m[i,j] == -100:
            colors[i,j] = np.array([1,1,1,1])
newC = np.concatenate((colors,colors),axis=1)
newC = np.concatenate((newC, newC),axis=0)
ax.imshow(newC,extent=[-5.9,5.9,-6.33,6.33])
#ax.imshow(colors)
cbar = fig.colorbar(p)
cbar.set_label("H(1/nm)")
ax.set_xlabel("X(nm)")
ax.set_ylabel("Y(nm)")
```
